# TASK-1: Table Booking and Online Delivery

Determine the percentage of restaurants that
offer table booking and online delivery

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Dataset(1).csv')
total_restaurants = data.shape[0]
restaurants_with_table_booking = data[data['Has Table booking'] == 'Yes'].shape[0]
restaurants_with_online_delivery = data[data['Has Online delivery'] == 'Yes'].shape[0]

percentage_of_table_booking = (restaurants_with_table_booking / total_restaurants) * 100
percentage_of_online_delivery = (restaurants_with_online_delivery / total_restaurants) * 100

print(f"Percentage of restaurants with Table Booking: {percentage_of_table_booking:.2f}%")
print(f"Percentage of restaurants with Online Delivery:{percentage_of_online_delivery:.2f}%")
```

Compare the average ratings of restaurants
with table booking and those without

```python
avg_rating_with_table_booking = data[data['Has Table booking'] == 'Yes']['Aggregate rating'].mean()
avg_rating_without_table_booking = data[data['Has Table booking'] == 'No']['Aggregate rating'].mean()

print("Average Rating with Table Booking:", avg_rating_with_table_booking)
print("Average Rating without Table Booking:", avg_rating_without_table_booking)
```

Analyze the availability of online delivery
among restaurants with different price ranges

```python
price_range_mapping = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}
data['Price Range'] = data['Price range'].map(price_range_mapping)
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Price Range', hue='Has Online delivery')
plt.title("Online Delivery Availability among Restaurants with Different Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.legend(title="Online Delivery")
plt.show()
```

# task-2:Price Range Analysis

Determine the most common price range
among all the restaurants

```python
most_common_price_range = data['Price range'].mode()[0]
print("Most Common Price Range:", most_common_price_range)
```

Calculate the average rating for each price
range

```python
average_rating_for_each_price_range = data.groupby('Price range')['Aggregate rating'].mean()
print("Average Rating for each Price Range:")
print(average_rating_for_each_price_range)
```

Identify the color that represents the highest
average rating among different price ranges.

```python
highest_avg_rating_color = data.groupby('Price range')['Rating color'].max()
print("Highest Average Rating Color by Price Range:")
print(highest_avg_rating_color)
```

```python
price_range_mapping = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}
data['Price Range'] = data['Price range'].map(price_range_mapping)
most_common_price_range = data['Price Range'].mode()[0]
print("Most Common Price Range:", most_common_price_range)
```
